```python
import numpy as np
import os, sys
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots()
scibert = np.load('scibert_pppl.npy')
bert = np.load('bert_pppl.npy')
roberta = np.load('roberta_pppl.npy')

scibert_nan = np.load('scibert_nan_filter_2_pppl.npy')

ax.plot(scibert, label='SciBERT')
ax.plot(scibert_nan, label='SciBERT-nan2', color='purple', linestyle='-')


ax.set_xlabel('Bin')
ax.set_ylabel('PPPL')

year_labels = {
    0 : '1879-1933',
    20 : '1993-1994',
    40 : '2006',
    60 : '2013',
    80 : '2018',
    99: '2023'
}

labels = [v for k, v in year_labels.items()]

ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(list(year_labels.keys()), labels)
ax2.set_xlabel('~Year')

ax.legend(loc='center left', bbox_to_anchor=(1.0,0.5))
fig.tight_layout()
fig.savefig('scibert.jpg', dpi=500)

ax.scatter(np.where(~np.isnan(bert))[0], bert[~np.isnan(bert)], color='red', label='BERT*')
ax.get_legend().remove()
ax.legend(loc='center left', bbox_to_anchor=(1.0,0.5))
fig.savefig('bert.jpg', dpi=500)

ax.scatter(np.where(~np.isnan(roberta))[0], roberta[~np.isnan(roberta)], color='green', label='RoBERTa*')
ax.get_legend().remove()
ax.legend(loc='center left', bbox_to_anchor=(1.0,0.5))
fig.savefig('roberta.jpg', dpi=500)
```
